**RENAMING THE COLUMNS**

```python
df.rename(columns = {'On a scale of 1 to 5, please indicate your level of confidence in providing healthcare services in a mobile setting, with 1 being the lowest and 5 being the highest:':'CONFIDENCE'}, inplace = True)
df.rename(columns = {'On a scale of 1 to 5, please indicate your level of familiarity with the protocols and guidelines for providing healthcare services in a mobile setting, with 1 being the lowest and 5 being the highest:':'FAMILIARITY WITH PROTOCOLS'}, inplace = True)
df.rename(columns = {'On a scale of 1 to 5, please indicate your level of comfort in adapting your healthcare skills to the unique challenges of delivering services in remote and underserved areas, with 1 being the lowest and 5 being the highest:':'COMFORT IN ADAPTING TO HEALTHCARE SKILLS'}, inplace = True)
df.rename(columns = {'On a scale of 1 to 5, please indicate whether you have received any training or professional development specifically related to delivering healthcare services in mobile units, with 1 being "No, I have not received any training" and 5 being "Yes, extensively":':'PRIOR TRAINING'}, inplace = True)
df.rename(columns = {'On a scale of 1 to 5, have you received proper training for using the Go Live app? (1 indicating no proper training received, 5 indicating strong agreement that proper training was received)':'TRAINING WITH GO LIVE APP'}, inplace = True)
df.rename(columns = {'On a scale of 1 to 5, how comfortable are you with using the Go Live app? (1 being very uncomfortable, 5 being very comfortable)':'COMFORT WITH THE APP'}, inplace = True)
df.rename(columns = {'On a scale of 1 to 5, would you like to receive additional training or support for using the Go Live app? (1 indicating strong disagreement, 5 indicating strong agreement)':'ADDITIONAL TRAINING REQUIREMENT'}, inplace = True)
```

**ANALYSING PARAMETERTS WITH RESPECT TO WORK EXPERIENCE**

```python
sns.barplot(data=df, x="Work Experience", y="CONFIDENCE")
```

```python
sns.barplot(data=df, x="Work Experience", y="FAMILIARITY WITH PROTOCOLS")
```

```python
sns.barplot(data=df, x="Work Experience", y="PRIOR TRAINING")
```

```python
sns.barplot(data=df, x="Work Experience", y="COMFORT IN ADAPTING TO HEALTHCARE SKILLS")
```

```python
sns.barplot(data=df, x="Work Experience", y="ADDITIONAL TRAINING REQUIREMENT")
```

**From the above graphs it can be analysed that:**

1. Very few people who had work experience between 3 to 5 years were totally confident in their work domain. Majority were equally confident irrespective of their work experience.
2. Few people with lesser work experience were also quite familiar with thr working protocols
3. Freshers were more adaptive to learning new skills
4. People with priot training were the ones with more work experince
5. People with lesser work experience were the ones who needed additional training which is natural. Hence our analysis is correct pertaining to the data provided
